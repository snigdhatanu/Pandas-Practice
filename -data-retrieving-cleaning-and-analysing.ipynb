## Genre counts per year (Extra Analysis)

```python
genres_rating_analysis = movies[["Year", "Genres"]]
```

```python
genre_dummies = genres_rating_analysis['Genres'].str.get_dummies(sep='|')
genres_rating_analysis = pd.concat([genres_rating_analysis, genre_dummies], axis=1)
genres_rating_analysis.drop('Genres', axis=1, inplace=True)
genres_rating_analysis
```

```python

genres_rating_analysis = pd.melt(genres_rating_analysis, id_vars=['Year'], var_name='Genre', value_name='value')
genres_rating_analysis
genre_year_sum = genres_rating_analysis.groupby(['Year', 'Genre']).sum()
genre_year_sum = genre_year_sum.sort_values(by=['Year', 'value'], ascending=False)
genre_year_sum
```

```python
genre_year_sum = genre_year_sum.reset_index()
genre_year_sum = genre_year_sum.loc[genre_year_sum.groupby('Year')['value'].idxmax()]
genre_year_sum = genre_year_sum.set_index(['Year', 'Genre'])
genre_year_sum = genre_year_sum.sort_values(by=['Year', 'value'], ascending=False)
```
